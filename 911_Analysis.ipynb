```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("911_Calls_for_Service_2022_New.csv")
```

```python
#print(type(data["Neighborhood"]))
neighborhoods = pd.unique(data["Neighborhood"])
call_reasons = pd.unique(data["description"])
Fed_hill = data[data['Neighborhood']=='Federal Hill']
Fed_hill = Fed_hill[Fed_hill['description'] != 'DIRECTED PATROL']
Fed_hill = Fed_hill[Fed_hill['description'] != 'Foot Patrol']


Fed_call_num = Fed_hill['description'].value_counts()

print(Fed_call_num[:51])

plt.pie(Fed_call_num.values, labels=Fed_call_num.keys(), autopct='%1.1f%%')
plt.show()
```
